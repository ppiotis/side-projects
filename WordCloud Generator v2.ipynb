```python
#import pandas as pd
#branchpack_jan = pd.read_fwf(r"C:\Users\PPIOTIS\Desktop\branch_pack.txt")

# open the file and read it into a variable alice_novel
branchpack_jan = open(r"C:\Users\PPIOTIS\Desktop\branch_pack.txt", 'r').read()
    
print ('File downloaded and saved!')
#print ('File downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
branchpack_wc = WordCloud(
    background_color='white',
    max_words=20000,
    stopwords=stopwords
)

# generate the word cloud
branchpack_wc.generate(branchpack_jan)
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_figwidth(40) # set width
fig.set_figheight(8) # set height

# display the word cloud
plt.imshow(branchpack_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object


branchpack_wc = WordCloud(background_color='white', max_words=2000, mask=shape, stopwords=stopwords)

# generate the word cloud
branchpack_wc.generate(branchpack_jan)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(20) # set width
fig.set_figheight(15) # set height

plt.imshow(branchpack_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
